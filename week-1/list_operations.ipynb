```python
# Accessing elements
lst = [10, 20, 30, 40]
print("First element:", lst[0])
print("Last element:", lst[-1])
```

```python
#Adding elements
lst = [10, 20, 30, 40]
lst.append(50)
print("After append:", lst)
lst.insert(2, 25)
print("After insert:", lst)
lst.extend([60, 70])
print("After extend:", lst)
```

```python
# Updating element
lst = [10, 20, 30, 40]
lst[0] = 100
print("After updating:", lst)
```

```python
# Removing elements
lst = [10, 20, 30, 40]
lst.remove(30)
print("After remove:", lst)
lst.pop()
print("After pop:", lst)
lst.pop(1)
print("After pop at index 1:", lst)
```

```python
# Searching
lst = [10, 20, 30, 40]
print("Index of 30:", lst.index(30))
print("Count of 20:", lst.count(20))
```

```python
# Looping
lst = [10, 20, 30, 40]
print("Elements in list:")
for item in lst:
    print(item)
```

```python
# Clearing list
lst = [10, 20, 30, 40]
lst.clear()
print("After clear:", lst)
```
